## ধাপ ৩ — নতুন Column তৈরি (Derived Columns)

````python
# বাড়িভাড়া ভাতা: মূল বেতনের ৪৫%
df['বাড়িভাড়া'] = (df['মূল_বেতন'] * 0.45).round(0).astype(int)

# চিকিৎসা ভাতা: সবার জন্য ১,৫০০ টাকা
df['চিকিৎসা_ভাতা'] = 1500

# মোট বেতন
df['মোট_বেতন'] = df['মূল_বেতন'] + df['বাড়িভাড়া'] + df['চিকিৎসা_ভাতা']

# চাকরির বয়স (বছরে)
আজকের_তারিখ = pd.Timestamp('2025-01-01')
df['চাকরির_বছর'] = ((আজকের_তারিখ - df['যোগদান_তারিখ']).dt.days / 365).round(1)

# বেতন স্তর নির্ধারণ
bins   = [0, 15000, 30000, 50000, 100000]
labels = ['নিম্ন (১৫ হাজারের নিচে)', 'মধ্যম (১৫-৩০ হাজার)',
          'উচ্চ-মধ্যম (৩০-৫০ হাজার)', 'উচ্চ (৫০ হাজারের উপরে)']
df['বেতন_স্তর'] = pd.cut(df['মূল_বেতন'], bins=bins, labels=labels)

print('নতুন column যোগ হয়েছে: বাড়িভাড়া, চিকিৎসা_ভাতা, মোট_বেতন, চাকরির_বছর, বেতন_স্তর')
df[['নাম','মূল_বেতন','বাড়িভাড়া','চিকিৎসা_ভাতা','মোট_বেতন','চাকরির_বছর']].head(5)
````

## ধাপ ৭ — GroupBy (বিভাগ অনুযায়ী বিশ্লেষণ)

````python
# বিভাগ অনুযায়ী গড় ও মোট বেতন
print('=== বিভাগভিত্তিক বেতন বিশ্লেষণ ===')
বিভাগ_বিশ্লেষণ = df.groupby('বিভাগ')['মূল_বেতন'].agg(
    কর্মকর্তা_সংখ্যা='count',
    মোট_বেতন='sum',
    গড়_বেতন='mean',
    সর্বোচ্চ='max',
    সর্বনিম্ন='min'
).round(0).sort_values('মোট_বেতন', ascending=False)

বিভাগ_বিশ্লেষণ
````

````python
# লিঙ্গভিত্তিক তুলনামূলক বিশ্লেষণ
print('=== লিঙ্গভিত্তিক বেতন তুলনা ===')
লিঙ্গ_বিশ্লেষণ = df.groupby('লিঙ্গ').agg(
    কর্মকর্তা_সংখ্যা=('নাম','count'),
    গড়_মূল_বেতন=('মূল_বেতন','mean'),
    গড়_মোট_বেতন=('মোট_বেতন','mean'),
    গড়_চাকরির_বছর=('চাকরির_বছর','mean')
).round(0)

লিঙ্গ_বিশ্লেষণ
````

````python
# বিভাগ ও লিঙ্গ — দুটো column দিয়ে groupby
print('=== বিভাগ ও লিঙ্গ অনুযায়ী বিশ্লেষণ ===')
df.groupby(['বিভাগ','লিঙ্গ'])['মূল_বেতন'].agg(['count','mean']).round(0)
````

## ধাপ ৮ — Pivot Table (সারসংক্ষেপ টেবিল)

````python
# বিভাগ vs লিঙ্গ — গড় বেতনের pivot table
print('=== Pivot Table: বিভাগ ও লিঙ্গ অনুযায়ী গড় মূল বেতন ===')
pivot = pd.pivot_table(
    df,
    values='মূল_বেতন',
    index='বিভাগ',
    columns='লিঙ্গ',
    aggfunc='mean',
    fill_value=0,
    margins=True,           # সর্বমোট সারি ও column
    margins_name='সর্বমোট'
).round(0)

pivot
````

````python
# বেতন স্তর vs বিভাগ — কর্মকর্তা সংখ্যার pivot
print('=== বেতন স্তর অনুযায়ী বিভাগভিত্তিক কর্মকর্তা সংখ্যা ===')
pd.pivot_table(
    df,
    values='নাম',
    index='বেতন_স্তর',
    columns='লিঙ্গ',
    aggfunc='count',
    fill_value=0,
    margins=True,
    margins_name='মোট'
)
````

## ধাপ ৯ — Apply ও Lambda (কাস্টম হিসাব)

````python
# বার্ষিক ইনক্রিমেন্ট হিসাব (সরকারি নিয়ম অনুযায়ী)
def ইনক্রিমেন্ট_হিসাব(row):
    """গ্রেড অনুযায়ী ইনক্রিমেন্ট হার নির্ধারণ"""
    if row['গ্রেড'] <= 5:
        return round(row['মূল_বেতন'] * 0.05)   # ৫%
    elif row['গ্রেড'] <= 10:
        return round(row['মূল_বেতন'] * 0.05)   # ৫%
    else:
        return round(row['মূল_বেতন'] * 0.05)   # ৫%

df['বার্ষিক_ইনক্রিমেন্ট'] = df.apply(ইনক্রিমেন্ট_হিসাব, axis=1)
df['পরবর্তী_বছর_বেতন']    = df['মূল_বেতন'] + df['বার্ষিক_ইনক্রিমেন্ট']

# Lambda দিয়ে সংক্ষিপ্ত পদ্ধতি
df['৩_বছর_পরের_বেতন'] = df['মূল_বেতন'].apply(lambda x: round(x * 1.05**3))

print('=== বার্ষিক ইনক্রিমেন্ট ও ভবিষ্যৎ বেতন প্রজেকশন ===')
df[['নাম','মূল_বেতন','বার্ষিক_ইনক্রিমেন্ট','পরবর্তী_বছর_বেতন','৩_বছর_পরের_বেতন']]
````

## ধাপ ১০ — Merge (দুটো DataFrame জোড়া লাগানো)

````python
# দ্বিতীয় DataFrame: প্রশিক্ষণ তথ্য
প্রশিক্ষণ_data = {
    'আইডি':          ['UZ-001','UZ-002','UZ-003','UZ-004','UZ-005',
                      'UZ-006','UZ-007','UZ-008'],
    'প্রশিক্ষণ':      ['e-Governance','Land Management','Python & AI','DGHS Training',
                      'Modern Agriculture','Education Policy','IBAS++','FP Training'],
    'প্রশিক্ষণ_ভাতা': [5000, 4000, 6000, 4500, 3500, 3500, 3000, 3500]
}
df_প্রশিক্ষণ = pd.DataFrame(প্রশিক্ষণ_data)

# Left merge: সব কর্মকর্তা থাকবে, প্রশিক্ষণ না থাকলে NaN
df_merge = pd.merge(df, df_প্রশিক্ষণ, on='আইডি', how='left')

# NaN এর জায়গায় ০
df_merge['প্রশিক্ষণ_ভাতা'] = df_merge['প্রশিক্ষণ_ভাতা'].fillna(0).astype(int)
df_merge['প্রশিক্ষণ']     = df_merge['প্রশিক্ষণ'].fillna('নেই')

# সর্বমোট বেতন = মোট বেতন + প্রশিক্ষণ ভাতা
df_merge['সর্বমোট_বেতন'] = df_merge['মোট_বেতন'] + df_merge['প্রশিক্ষণ_ভাতা']

print('=== Merge এর পরে ডেটা ===')
df_merge[['নাম','মোট_বেতন','প্রশিক্ষণ','প্রশিক্ষণ_ভাতা','সর্বমোট_বেতন']]
````

## ধাপ ১৩ — চূড়ান্ত রিপোর্ট ও Excel Export

````python
# চূড়ান্ত বেতন বিবরণী DataFrame
চূড়ান্ত_রিপোর্ট = df[[
    'আইডি','নাম','পদবি','বিভাগ','গ্রেড','লিঙ্গ',
    'যোগদান_তারিখ','চাকরির_বছর',
    'মূল_বেতন','বাড়িভাড়া','চিকিৎসা_ভাতা','মোট_বেতন',
    'বার্ষিক_ইনক্রিমেন্ট','পরবর্তী_বছর_বেতন'
]].copy()

# সারসংক্ষেপ সারি যোগ
সারসংক্ষেপ = pd.DataFrame([{
    'আইডি': '---',
    'নাম': 'সর্বমোট',
    'মূল_বেতন':           চূড়ান্ত_রিপোর্ট['মূল_বেতন'].sum(),
    'বাড়িভাড়া':           চূড়ান্ত_রিপোর্ট['বাড়িভাড়া'].sum(),
    'চিকিৎসা_ভাতা':       চূড়ান্ত_রিপোর্ট['চিকিৎসা_ভাতা'].sum(),
    'মোট_বেতন':           চূড়ান্ত_রিপোর্ট['মোট_বেতন'].sum(),
    'বার্ষিক_ইনক্রিমেন্ট': চূড়ান্ত_রিপোর্ট['বার্ষিক_ইনক্রিমেন্ট'].sum(),
    'পরবর্তী_বছর_বেতন':   চূড়ান্ত_রিপোর্ট['পরবর্তী_বছর_বেতন'].sum(),
}])

চূড়ান্ত_রিপোর্ট = pd.concat([চূড়ান্ত_রিপোর্ট, সারসংক্ষেপ], ignore_index=True)
print('=== চূড়ান্ত বেতন বিবরণী ===')
চূড়ান্ত_রিপোর্ট
````

````python
# Excel-এ export — একাধিক sheet-এ
try:
    with pd.ExcelWriter('উপজেলা_বেতন_রিপোর্ট.xlsx', engine='openpyxl') as writer:

        # Sheet 1: সম্পূর্ণ বিবরণী
        চূড়ান্ত_রিপোর্ট.to_excel(writer, sheet_name='বেতন বিবরণী', index=False)

        # Sheet 2: বিভাগভিত্তিক সারসংক্ষেপ
        বিভাগ_বিশ্লেষণ.to_excel(writer, sheet_name='বিভাগভিত্তিক')

        # Sheet 3: Pivot Table
        pivot.to_excel(writer, sheet_name='Pivot Table')

        # Sheet 4: ভবিষ্যৎ প্রজেকশন
        df[['নাম','মূল_বেতন','পরবর্তী_বছর_বেতন','৩_বছর_পরের_বেতন']].to_excel(
            writer, sheet_name='বেতন প্রজেকশন', index=False
        )

    print('✓ উপজেলা_বেতন_রিপোর্ট.xlsx সফলভাবে তৈরি হয়েছে!')
    print('  Sheet সমূহ: বেতন বিবরণী, বিভাগভিত্তিক, Pivot Table, বেতন প্রজেকশন')
except ImportError:
    print('openpyxl নেই। চালাও: pip install openpyxl')
    df.to_csv('উপজেলা_বেতন_রিপোর্ট.csv', index=False, encoding='utf-8-sig')
    print('✓ CSV হিসেবে সংরক্ষিত হয়েছে।')
````

## ধাপ ১৪ — Text-based ভিজুয়ালাইজেশন (Matplotlib ছাড়া)

````python
print('=' * 60)
print('       উপজেলা বেতন বিশ্লেষণ — সারসংক্ষেপ ড্যাশবোর্ড')
print('=' * 60)

মোট = df['মোট_বেতন'].sum()
print(f'\n  মোট কর্মকর্তা-কর্মচারী : {len(df):>5} জন')
print(f'  মাসিক মোট বেতন বিল     : ৳{মোট:>10,.0f}')
print(f'  বার্ষিক মোট বেতন বিল   : ৳{মোট*12:>10,.0f}')
print(f'  গড় মোট বেতন            : ৳{df["মোট_বেতন"].mean():>10,.0f}')

print('\n--- বিভাগভিত্তিক বেতন বিল (বার চার্ট) ---')
বিভাগ_মোট = df.groupby('বিভাগ')['মোট_বেতন'].sum().sort_values(ascending=False)
সর্বোচ্চ = বিভাগ_মোট.max()
for বিভাগ, মোট_ব in বিভাগ_মোট.items():
    বার_দৈর্ঘ্য = int((মোট_ব / সর্বোচ্চ) * 30)
    print(f'  {বিভাগ:<12} {"█" * বার_দৈর্ঘ্য} ৳{মোট_ব:,.0f}')

print('\n--- বেতন স্তর বিতরণ ---')
স্তর_গণনা = df['বেতন_স্তর'].value_counts()
for স্তর, গণনা in স্তর_গণনা.items():
    বার = '●' * গণনা
    print(f'  {str(স্তর)[:30]:<32} {বার} ({গণনা} জন)')

print('\n--- লিঙ্গভিত্তিক তথ্য ---')
for লিঙ্গ, গ্রুপ in df.groupby('লিঙ্গ'):
    print(f'  {লিঙ্গ}: {len(গ্রুপ)} জন, গড় বেতন ৳{গ্রুপ["মূল_বেতন"].mean():,.0f}')

print('\n' + '=' * 60)
print(f'  সর্বোচ্চ বেতনধারী: {df.loc[df["মূল_বেতন"].idxmax(), "নাম"]} (৳{df["মূল_বেতন"].max():,})')
print(f'  সর্বনিম্ন বেতনধারী: {df.loc[df["মূল_বেতন"].idxmin(), "নাম"]} (৳{df["মূল_বেতন"].min():,})')
print('=' * 60)
````
