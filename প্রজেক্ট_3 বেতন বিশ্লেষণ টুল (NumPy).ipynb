```python
import numpy as np

# নলডাঙ্গা উপজেলার ১০ জন কর্মচারীর তথ্য: [মূল_বেতন, গ্রেড, অভিজ্ঞতা(বছর)]
karmochari_data = np.array([
    [22000, 9, 3],
    [35000, 7, 8],
    [28000, 8, 5],
    [41000, 6, 12],
    [19000, 10, 1],
    [52000, 5, 15],
    [31000, 7, 6],
    [24000, 9, 2],
    [38000, 6, 10],
    [27000, 7, 3]
])

nam = ["আজম", "করিম", "রহিমা", "সালাম", "ফাতেমা", 
       "জাহিদ", "নাসরিন", "রফিক", "শাহিনা", "মামুন"]

beton = karmochari_data[:, 0]   # প্রথম কলাম = মূল বেতন

print("=== মূল পরিসংখ্যান ===")
print(f"মোট কর্মচারী: {len(beton)}")
print(f"গড় বেতন: {beton.mean():.0f} টাকা")
print(f"মধ্যমা বেতন: {np.median(beton):.0f} টাকা")
print(f"সর্বনিম্ন: {beton.min()} টাকা")
print(f"সর্বোচ্চ: {beton.max()} টাকা")
print(f"স্ট্যান্ডার্ড ডেভিয়েশন: {beton.std():.0f} টাকা")
```

```python
oviggota = karmochari_data[:, 2]   # তৃতীয় কলাম = অভিজ্ঞতা

# boolean mask তৈরি — কোন শর্ত কাদের জন্য সত্য
kom_oviggota = oviggota < 5

# np.where দিয়ে শর্তভিত্তিক broadcasting
increment_rate = np.where(kom_oviggota, 1.10, 1.07)
notun_beton = beton * increment_rate

print("=== ইনক্রিমেন্ট পর নতুন বেতন ===")
for i in range(len(nam)):
    print(f"{nam[i]:10s} | অভিজ্ঞতা: {oviggota[i]:2d} বছর | "
          f"আগের: {beton[i]:6d} | নতুন: {notun_beton[i]:8.0f}")
```

```python
# প্রতিটি কর্মচারীর গ্রেড অনুযায়ী one-hot মতো ভাগ করা সহজ পদ্ধতি না করে,
# সরাসরি গ্রেড-ভিত্তিক বোনাস ডিকশনারি ব্যবহার করি
grade = karmochari_data[:, 1]   # দ্বিতীয় কলাম = গ্রেড

# গ্রেড যত কম (১ এর কাছে), পদমর্যাদা তত উঁচু, বোনাস তত বেশি
bonus_rate = (11 - grade) * 500   # গ্রেড ৫ -> বোনাস ৩০০০, গ্রেড ১০ -> বোনাস ৫০০
mot_bonus = bonus_rate.sum()

print(f"মোট বোনাস বাজেট: {mot_bonus:.0f} টাকা")
print(f"মোট বেতন বাজেট (ইনক্রিমেন্ট পর): {notun_beton.sum():.0f} টাকা")
print(f"সর্বমোট মাসিক বাজেট: {(notun_beton.sum() + mot_bonus):.0f} টাকা")
```

```python
gor = beton.mean()

beshi_betonvogi = [nam[i] for i in range(len(nam)) if beton[i] > gor]
kom_betonvogi = [nam[i] for i in range(len(nam)) if beton[i] <= gor]

print(f"\nগড় বেতনের ({gor:.0f} টাকা) চেয়ে বেশি পান যারা:")
print(", ".join(beshi_betonvogi))

print(f"\nগড় বেতনের সমান বা কম পান যারা:")
print(", ".join(kom_betonvogi))

# percentile বিশ্লেষণ
print(f"\n২৫তম পার্সেন্টাইল: {np.percentile(beton, 25):.0f} টাকা")
print(f"৭৫তম পার্সেন্টাইল: {np.percentile(beton, 75):.0f} টাকা")
```
