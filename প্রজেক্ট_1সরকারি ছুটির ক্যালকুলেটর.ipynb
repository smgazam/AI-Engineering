```python
# ২০২৬ সালের কিছু সরকারি ছুটি (mas, din) আকারে
sorkari_chuti = {
    (1, 21): "শহীদ দিবস (একুশে ফেব্রুয়ারি প্রস্তুতি)",
    (2, 21): "আন্তর্জাতিক মাতৃভাষা দিবস",
    (3, 26): "স্বাধীনতা দিবস",
    (4, 14): "পহেলা বৈশাখ",
    (5, 1): "মে দিবস",
    (8, 15): "জাতীয় শোক দিবস",
    (12, 16): "বিজয় দিবস",
    (12, 25): "বড়দিন"
}

print("মোট ছুটির সংখ্যা:", len(sorkari_chuti))
```

```python
def chuti_check(mas, din):
    """একটি নির্দিষ্ট তারিখ ছুটি কিনা যাচাই করে"""
    key = (mas, din)
    if key in sorkari_chuti:
        return True, sorkari_chuti[key]
    else:
        return False, None

# পরীক্ষা করি
mas = 3
din = 26
ache, naam = chuti_check(mas, din)

if ache:
    print(f"{mas}/{din} তারিখে ছুটি আছে: {naam}")
else:
    print(f"{mas}/{din} তারিখে কোনো ছুটি নেই")
```

```python
def porer_chuti(current_mas, current_din):
    """বর্তমান তারিখের পর প্রথম ছুটি খুঁজে বের করে"""
    sorted_dates = sorted(sorkari_chuti.keys())
    
    for mas, din in sorted_dates:
        if mas > current_mas or (mas == current_mas and din > current_din):
            return mas, din, sorkari_chuti[(mas, din)]
    
    # যদি বছরের আর কোনো ছুটি না থাকে, প্রথমটিই দেখাও
    first = sorted_dates[0]
    return first[0], first[1], sorkari_chuti[first]

# পরীক্ষা
mas, din, naam = porer_chuti(5, 10)
print(f"পরবর্তী ছুটি: {mas}/{din} — {naam}")
```

```python
def main():
    print("=== সরকারি ছুটি ক্যালকুলেটর ===\n")
    
    mas = int(input("মাস লিখুন (1-12): "))
    din = int(input("দিন লিখুন (1-31): "))
    
    ache, naam = chuti_check(mas, din)
    
    if ache:
        print(f"\n✓ {mas}/{din} তারিখে ছুটি: {naam}")
    else:
        print(f"\n✗ {mas}/{din} তারিখে অফিস খোলা")
        n_mas, n_din, n_naam = porer_chuti(mas, din)
        print(f"  → পরবর্তী ছুটি: {n_mas}/{n_din} ({n_naam})")

main()
```
