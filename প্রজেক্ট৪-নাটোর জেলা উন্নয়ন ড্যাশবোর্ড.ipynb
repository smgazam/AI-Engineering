```python
import pandas as pd
import numpy as np

# নাটোর জেলার ৬টি উপজেলার উন্নয়ন সূচক
data = {
    "উপজেলা": ["নলডাঙ্গা", "সিংড়া", "গুরুদাসপুর",
                "বড়াইগ্রাম", "লালপুর", "বাগাতিপাড়া"],
    "জনসংখ্যা": [142000, 285000, 198000, 176000, 210000, 134000],
    "সাক্ষরতার_হার": [68.5, 72.3, 65.8, 70.1, 69.4, 67.2],
    "ই_সেবা_কেন্দ্র": [3, 5, 4, 3, 4, 2],
    "মাসিক_আবেদন": [340, 820, 510, 390, 480, 260],
    "বিদ্যুৎ_কভারেজ": [88.2, 92.5, 85.6, 89.3, 87.1, 83.4],
    "রাস্তা_কিমি": [145, 312, 198, 167, 225, 118]
}

df = pd.DataFrame(data)
df.to_csv("natore_unnayan.csv", index=False, encoding="utf-8")
print("ডেটাসেট তৈরি হয়েছে!")
print(df)
```

```python
df = pd.read_csv("natore_unnayan.csv", encoding="utf-8")

# প্রতি ১০,০০০ জনের জন্য মাসিক আবেদন হার (তুলনামূলক সূচক)
df["আবেদন_হার_প্রতি_১০হাজার"] = (
    df["মাসিক_আবেদন"] / df["জনসংখ্যা"] * 10000
).round(2)

# ই-সেবা প্রতি কেন্দ্রের গড় আবেদন
df["গড়_আবেদন_প্রতি_কেন্দ্র"] = (
    df["মাসিক_আবেদন"] / df["ই_সেবা_কেন্দ্র"]
).round(0)

print("=== উন্নয়ন সূচক ===")
print(df[["উপজেলা", "আবেদন_হার_প্রতি_১০হাজার",
          "গড়_আবেদন_প্রতি_কেন্দ্র"]].sort_values(
    "আবেদন_হার_প্রতি_১০হাজার", ascending=False
))

# সাক্ষরতা ৬৮%-এর নিচে উপজেলা চিহ্নিত করা
kom_shakhhorota = df[df["সাক্ষরতার_হার"] < 68]
print("\n=== সাক্ষরতা ৬৮%-এর নিচে ===")
print(kom_shakhhorota[["উপজেলা", "সাক্ষরতার_হার", "জনসংখ্যা"]])
```

```python
# ই-সেবা কেন্দ্র সংখ্যা অনুযায়ী গ্রুপ বিশ্লেষণ
group_analysis = df.groupby("ই_সেবা_কেন্দ্র").agg(
    উপজেলা_সংখ্যা=("উপজেলা", "count"),
    গড়_জনসংখ্যা=("জনসংখ্যা", "mean"),
    গড়_মাসিক_আবেদন=("মাসিক_আবেদন", "mean"),
    গড়_সাক্ষরতা=("সাক্ষরতার_হার", "mean")
).round(1)

print("=== ই-সেবা কেন্দ্র সংখ্যাভিত্তিক বিশ্লেষণ ===")
print(group_analysis)

# জেলার সর্বমোট ও গড়
print("\n=== জেলা সারসংক্ষেপ ===")
print(f"মোট জনসংখ্যা: {df['জনসংখ্যা'].sum():,}")
print(f"গড় সাক্ষরতার হার: {df['সাক্ষরতার_হার'].mean():.1f}%")
print(f"মোট মাসিক আবেদন: {df['মাসিক_আবেদন'].sum():,}")
print(f"গড় বিদ্যুৎ কভারেজ: {df['বিদ্যুৎ_কভারেজ'].mean():.1f}%")
```

```python
# ICT কর্মকর্তার তথ্য
ict_officers = pd.DataFrame({
    "উপজেলা": ["নলডাঙ্গা", "সিংড়া", "গুরুদাসপুর",
                "বড়াইগ্রাম", "লালপুর", "বাগাতিপাড়া"],
    "ICT_কর্মকর্তা": ["আজম", "করিম", "রহিমা",
                       "সালাম", "ফাতেমা", "জাহিদ"],
    "অভিজ্ঞতা_বছর": [7, 12, 5, 9, 3, 8]
})

# merge করা
full_df = df.merge(ict_officers, on="উপজেলা", how="left")

# চূড়ান্ত রিপোর্ট
print("=== নাটোর জেলা ICT ড্যাশবোর্ড ===")
report = full_df[[
    "উপজেলা", "জনসংখ্যা", "সাক্ষরতার_হার",
    "মাসিক_আবেদন", "আবেদন_হার_প্রতি_১০হাজার",
    "ICT_কর্মকর্তা", "অভিজ্ঞতা_বছর"
]].sort_values("আবেদন_হার_প্রতি_১০হাজার", ascending=False)

print(report.to_string(index=False))

# চূড়ান্ত ফাইল সেভ
report.to_csv("natore_ict_dashboard.csv",
              index=False, encoding="utf-8")
print("\nচূড়ান্ত রিপোর্ট সেভ হয়েছে: natore_ict_dashboard.csv")
```
